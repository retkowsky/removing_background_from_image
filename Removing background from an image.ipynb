```python
from rembg import remove
import cv2
import matplotlib.pyplot as plt
```

```python
def remove_bkg(myimage):
    
    outputfile = 'result_' + myimage
    img = cv2.imread(myimage) # Reading the image
    result = remove(img) # Removing the background
    cv2.imwrite(outputfile, result) # Saving the file
    return myimage, outputfile
```

```python
def display_results(myimage, resultimage):
    
    image1 = plt.imread(myimage)
    image2 = plt.imread(resultimage)
    
    NUM_ROWS = 1
    IMGS_IN_ROW = 2 # 2 images side by side
    f, ax = plt.subplots(NUM_ROWS, IMGS_IN_ROW, figsize = (12, 4))
    ax[0].imshow(image1)
    ax[1].imshow(image2)
    ax[0].set_title('Initial image')
    ax[1].set_title('Image without the background')
    title = 'Removing the background from an image'
    f.suptitle(title, fontsize = 15)
    plt.tight_layout()
    plt.savefig('sidebyside_' + myimage, dpi = 300) # Saving the side by side image
    plt.show()
```

```python
def removing_background(myimage):
    
    results = remove_bkg(myimage) # Main program
    display_results(results[0], results[1]) # Side by side images
```

```python
removing_background('queen.jpg')
```
